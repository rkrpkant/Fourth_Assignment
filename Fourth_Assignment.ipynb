```python
# Q3. Create a dataset representing categories and their corresponding values. Compare different categories
# based on numerical values.

# Ans.

import pandas as pd

# Sample data (replace with your actual data)
data = {'Category': ['A', 'A', 'B', 'B', 'C', 'C', 'A', 'B'],
        'Value': [10, 15, 20, 25, 30, 35, 12, 22]}
df = pd.DataFrame(data)

# Group by category and calculate statistics
category_stats = df.groupby('Category')['Value'].agg(['mean', 'median', 'std'])

print(category_stats)


# Compare categories (example: mean values)
category_means = df.groupby('Category')['Value'].mean()

# Find the category with the highest mean
highest_mean_category = category_means.idxmax()
print(f"The category with the highest mean value is: {highest_mean_category}")

# Compare pairs of categories
# Example: Compare categories A and B
mean_A = category_means['A']
mean_B = category_means['B']

if mean_A > mean_B:
  print("Category A has a higher mean value than Category B.")
elif mean_A < mean_B:
  print("Category B has a higher mean value than Category A.")
else:
  print("Category A and Category B have the same mean value.")
```

```python
# Q5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
# heatmap.

# Ans.

# Generate a synthetic dataset with correlated features
np.random.seed(0)
n_samples = 100
x1 = np.random.rand(n_samples)
x2 = 0.8 * x1 + 0.1 * np.random.randn(n_samples)
x3 = 0.5 * x1 + 0.6 * x2 + 0.1 * np.random.randn(n_samples)

# Create a Pandas DataFrame
df = pd.DataFrame({'x1': x1, 'x2': x2, 'x3': x3})

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
# Q3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and
# days.
#  np.random.seed(20)
#  data = {
#  'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
#  'Day': np.random.choice(range(1, 31), 100),
#  'Sales': np.random.randint(1000, 5000, 100)
#  }
#  df = pd.DataFrame(data)

# Sales Data Heatmap (This code was already present, but is included for completeness)

# Ans.

np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df_sales = pd.DataFrame(data)

sales_pivot = df_sales.pivot_table(index='Month', columns='Day', values='Sales', aggfunc='sum')

plt.figure(figsize=(10, 6))
sns.heatmap(sales_pivot, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title('Sales Variation Across Months and Days')
plt.xlabel('Day')
plt.ylabel('Month')
plt.show()
```

```python
# Q4. Using the given x and y data, generate a 3D surface plot to visualize the function
# x = np.linspace(-5, 5, 100)
#  y = np.linspace(-5, 5, 100)
#  x, y = np.meshgrid(x, y)
#  z = np.sin(np.sqrt(x**2 + y**2))
#  data = {
#  'X': x.flatten(),
#  'Y': y.flatten(),
#  'Z': z.flatten()
#  }
#  df = pd.DataFrame(data)

# Ans.

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot_trisurf(df['X'], df['Y'], df['Z'], cmap='viridis')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot')

plt.show()
```

```python
# Q5. Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (x
# axis), and bubble size proportional to the population.
#  np.random.seed(25)
#  data = {
#  'Country': ['USA', 'Canada', 'UK',
# 'Germany', 'France'],
#  'Population':
# np.random.randint(100, 1000, 5),
#  'GDP': np.random.randint(500, 2000,
# 5)
#  }
#  df = pd.DataFrame(data)

# Ans.

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

plt.figure(figsize=(8, 6))
plt.scatter(df['GDP'], df['Population'], s=df['Population'] * 10, alpha=0.6)  # Adjust bubble size as needed

for i, country in enumerate(df['Country']):
    plt.annotate(country, (df['GDP'][i], df['Population'][i]), fontsize=10)

plt.xlabel("GDP")
plt.ylabel("Population")
plt.title("Bubble Chart: Country Population vs GDP")
plt.show()
```

```python
                                                            # BOKEH ASSIGNMENT
# Q1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

# Ans.

import numpy as np
from bokeh.plotting import figure, show, output_notebook

# Output the plot in the notebook
output_notebook()

# Create x-values from 0 to 10
x = np.linspace(0, 10, 100)

# Calculate y-values as the sine of x
y = np.sin(x)

# Create a new plot with a title and axis labels
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="Sine", line_width=2)

# Show the plot
show(p)
```

```python
# Q2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
# markers based on the 'sizes' and 'colors' columns.


# Ans.

import pandas as pd
import numpy as np
from bokeh.plotting import figure, show, output_notebook

# Output the plot in the notebook
output_notebook()

# Create a sample DataFrame (replace with your actual data)
n = 50
df = pd.DataFrame({
    'x': np.random.rand(n) * 10,
    'y': np.random.rand(n) * 10,
    'sizes': np.random.randint(1, 20, n),
    'colors': np.random.choice(['red', 'green', 'blue'], n)
})

# Create a Bokeh figure
p = figure(title="Scatter Plot with Sizes and Colors",
           x_axis_label='x', y_axis_label='y',
           tools="pan,wheel_zoom,reset,save")

# Add circles to the figure with sizes and colors from the DataFrame
p.circle(x='x', y='y', size='sizes', color='colors', source=df)

# Show the plot
show(p)
```

```python
# Q4. Create a Bokeh histogram to visualize the distribution of the given data.
#  data_hist = np.random.randn(1000)
#  hist, edges = np.histogram(data_hist, bins=30)


# Ans.

# Output the plot in the notebook
output_notebook()
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a histogram
p = figure(title="Histogram", x_axis_label='Value', y_axis_label='Frequency')

# Add a quad glyph for the histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color="navy", line_color="white", alpha=0.5)

# Show the plot
show(p)
```
